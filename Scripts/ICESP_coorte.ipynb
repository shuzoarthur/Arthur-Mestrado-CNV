```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
icesp_coorte = pd.read_csv('/content/Mestrado/Dados/ICESP_coorte/RetratosDaMama-Arthur_DATA_LABELS_2024-11-11_1423.csv', sep=';')
icesp_coorte.columns
```

```python
import re

# Define the pattern to match
pattern = r'^Genes com achados reportáveis.+'

# Get a list of columns that match the pattern
columns_to_remove = [col for col in icesp_coorte.columns if re.match(pattern, col)]
lista = ['Repeat Instrument', 'Repeat Instance', 'Raça declarada', 'IMC',
       'Possui histórico familiar de câncer?', 'Hábito de fumar',
       'Ingere bebidas alcóolicas regularmente?',]
columns_to_remove.extend(lista)

# Remove the columns
icesp_coorte = icesp_coorte.drop(columns=columns_to_remove)
```
